```python
import pandas as pd
import numpy as np
import os
```

```python
import kagglehub

# Download kar raha hu yaha :hahaha
path = kagglehub.dataset_download("emrullahdemirhan/titaniccsv")


data = pd.read_csv(os.path.join(path, 'titanic.csv'))


new = data[['Survived','Ticket', 'Cabin']]

print(new.head())
```

```python
df = pd.DataFrame(new)
df.head()
```

```python
df.info()
```

```python
df['Survived'].unique()
```

```python
df['Ticket'].unique()
```

yaha se dono ko alag alag extract kr do


```python
df['Cabin_num'] =  df['Cabin'].str.extract('(\d+)')
df['Cabin_cat'] =  df['Cabin'].str[0]
df.head()
```

```python
df['Cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['Ticket'].dtype
```

```python
df['Ticket_num'] = df['Ticket'].str.extract(r'(\d+)').astype(float)
```

```python
df['Ticket_cat'] = df['Ticket'].str[0]
```
